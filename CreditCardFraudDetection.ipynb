# Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading a dataset to pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# Print first 5 rows of the data set
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# Dataset Information
credit_card_data.info()
```

```python
# Checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit and fraudulent transactions
# where 0 = legit and 1 = fraudulent

credit_card_data['Class'].value_counts()
```

This data set is highly unbalanced

0 == legit transactions
1 == fraudulent transactions


```python
# separating the data for analysis

legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistacal measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

Under - Sampling

Build a sample dataset containing similar distributions of normal transactions and fradulunt transactions

# no of fradulunt trans = 487 equaling it norma trans

```python
legit_sample = legit.sample(n=487)
```

Concatinating Two DataFrames

```python
new_dataset = pd.concat([legit_sample,fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features(columns) and Targets(Labels 0&1)

```python
X = new_dataset.drop(columns="Class", axis = 1)
Y = new_dataset["Class"]
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Traning set and Test set

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify = Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# Training the logistic regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy score

```python
# Accuracy on traing data
X_train_prediction = model.predict(X_train)
traning_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on Training data:", traning_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on Test data:", test_data_accuracy)
```

Overfitting Vs Underfitting

If the accuracy of our training data is 94% and accuracy of test data is 50% then it is Overfitting(doestn't get generalised learning)
If the accuracy of our training data is 50% and accuracy of test data is 90% then it is Underfitting(over generalised learning)
